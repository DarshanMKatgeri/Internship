# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
    a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
    b) Top 10 ODI Batsmen along with the records of their team and rating.
    c) Top 10 ODI bowlers along with the records of their team and rating.

```python
#importing basic libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
#Sending request to the webpage
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
page_teams = requests.get(url)
page_teams
```

```python
#Instantiating
soup = bs(page_teams.content)
print(soup.prettify())
```

# variable names:
       a)Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
    teams          =  list of all teams
    matches_banner =  matches won by rank-1 team
    points_banner  =  points of rank-1 team
    ratings_banner =  ratings of rank-1 team
    ------------------------------------------------------------------------------
    matches_all =  list of matches won by teams ranked from 2 to 10
    points_all  =  list of points of teams ranked from 2 to 10
    ratings_all =  list of ratings of teams ranked from 2 to 10    
    matches_points = list of matches and points of teams ranked from 2 to 10 (has to be seperated)
    ------------------------------------------------------------------------------
       b) Top 10 ODI Batsmen along with the records of their team and rating.
    batsman_banner_name    = name of rank-1 batsman
    batsman_banner_country = country of rank-1 batsman
    batsman_banner_rating  = rating of rank-1 batsman
    ------------------------------------------------------------------------------
    batsman_all_name      = names of batsman starting from rank-2
    batsman_all_country   = country of batsman starting from rank-2
    batsman_all_rating    = ratings of batsman starting from rank-2
    ------------------------------------------------------------------------------
    batsman_banner_df  = rank-1 batsman dataframe
    batsman_all = top 100 batsman dataframe
    

```python
teams = []
for i in soup.find_all('span', class_="u-hide-phablet"):
    teams.append(i.text)
teams
teams = pd.DataFrame(teams)
teams.columns = ['team']
index = np.arange(2,len(teams)+2)
teams_df = pd.DataFrame(teams, index=index)
teams_df
```

# ---------------------------------------------------------------------------------------------------

# Extracting matches, point and rating info from banner of html file. In this site banner holds info of highest ranked team. Banner(Team with rank-1) info is extracted first, later info about other teams is extracted. Later dataframes are formed.

```python
matches_banner = []
for j in soup.find_all('td', class_="rankings-block__banner--matches"):
    matches_banner.append(j.text)
matches_banner
m_df = pd.DataFrame(matches_banner)
m_df
```

```python
points_banner = []
for k in soup.find_all('td',class_="rankings-block__banner--points"):
    points_banner.append(k.text)
points_banner
p_df = pd.DataFrame(points_banner)
p_df
```

```python
ratings_banner = []
for l in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    ratings_banner.append(l.text.replace('\n','').replace(' ',''))
ratings_banner
r_df = pd.DataFrame(ratings_banner)
r_df
```

```python
df= pd.concat([m_df, p_df, r_df],axis=1)
df.columns = ['matches','points','rating']
#index = np.arange(1, len(df))
df
```

```python
matches_points = []
for mp in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches_points.append(mp.text)
matches_points
```

```python
matches = []
points = []
for i in range(len(matches_points)):
    if i%2 == 0:
        matches.append(matches_points[i])
    else:
        points.append(matches_points[i])

print("Matches: ", matches)
print("points: ", points)
```

```python
index = np.arange(2, len(matches)+2)
matches_df = pd.DataFrame(matches, index=index)
matches_df.columns =['matches']
matches_df
```

```python
index = np.arange(2, len(points)+2)
points_df = pd.DataFrame(points, index=index)
points_df.columns =['points']
points_df
```

```python
ratings_all = []
for m in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    ratings_all.append(m.text)
ratings_all
index = np.arange(2, len(ratings_all)+2)
ratings_all = pd.DataFrame(ratings_all, index=index)
ratings_all.columns = ['rating']
ratings_all
```

```python
df_final = pd.concat([matches_df, points_df, ratings_all], axis=1)
df_final
```

```python
frm = pd.concat([df, df_final], axis=0)
index = np.
frm
```

# b) Top 10 ODI Batsmen along with the records of their team and rating.

```python
#Sending request to the webpage
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting'
page_batsman = requests.get(url)
page_batsman
```

```python
soup_batsman = bs(page_batsman.content)
soup_batsman
```

# name, team and rating of rank 1 batsman

```python
batsman_banner_name = []
for bbn in soup_batsman.find_all('div', class_="rankings-block__banner--name-large"):
    batsman_banner_name.append(bbn.text)
batsman_banner_name
```

```python
batsman_banner_country = []
for bbc in soup_batsman.find_all('div',class_="rankings-block__banner--nationality"):
    batsman_banner_country.append(bbc.text.replace('\n','').replace(' ', ''))
batsman_banner_country
```

```python
batsman_banner_rating = []
for bbr in soup_batsman.find_all('div',class_="rankings-block__banner--rating"):
    batsman_banner_rating.append(bbr.text)
batsman_banner_rating
```

# creating dataframe for rank-1 batsman

```python
batsman_banner_name_df = pd.DataFrame(batsman_banner_name, index=[1])
batsman_banner_country_df = pd.DataFrame(batsman_banner_country, index=[1])
batsman_banner_rating_df = pd.DataFrame(batsman_banner_rating, index=[1])
```

```python
batsman_banner_name_df.columns = ['name']
batsman_banner_name_df
```

```python
batsman_banner_country_df.columns = ['country']
batsman_banner_country_df
```

```python
batsman_banner_rating_df.columns = ['rating']
batsman_banner_rating_df
```

```python
batsman_banner_df = pd.concat([batsman_banner_name_df, batsman_banner_country_df, batsman_banner_rating_df],axis=1)
batsman_banner_df
```

# name, team and rating of rank 2 onwards batsman

```python
batsman_all_country = []
for bac in soup_batsman.find_all('span',class_="table-body__logo-text"):
    batsman_all_country.append(bac.text)
print(batsman_all_country)
```

# creating dataframe for rank-2 onwards batsman

```python
batsman_all_name_df = pd.DataFrame(batsman_all_name)
index = np.arange(2,len(batsman_all_name_df)+2)
batsman_all_name_df = pd.DataFrame(batsman_all_name,index=index)
batsman_all_name_df
```

```python
batsman_all = pd.concat([batsman_all_name_df, batsman_all_country_df, batsman_all_rating_df], axis=1)
batsman_all.columns = ['name','country', 'rating']
batsman_all
```
