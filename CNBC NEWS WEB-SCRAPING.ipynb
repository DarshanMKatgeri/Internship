# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
    i) Headline
    ii) Time
    iii) News Link

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

# Requesting the content
cnbc = requests.get('https://www.cnbc.com/world/?region=world')
print(cnbc)

# Displaying the content
news = bs(cnbc.content)
#print(news.prettify())
```

```python
headlines =[]
for i in news.find_all('a', class_="LatestNews-headline"):
    headlines.append(i.text)
#headlines
```

```python
time = []
for j in news.find_all('time', class_="LatestNews-timestamp"):
    time.append(j.text)
#time
```

```python
link = []
for k in news.find_all('a', class_="LatestNews-headline"):
    link.append(k.get('href'))
#link
```

# creating dataframes

```python
headlines_df = pd.DataFrame(headlines)
#headlines_df
```

```python
time_df = pd.DataFrame(time)
#time_df
```

```python
link_df = pd.DataFrame(link)
#link_df
```

# final required headline time and link dataframe

```python
cnbc_df = pd.concat([headlines_df,time_df,link_df], axis=1)
cnbc_df.columns = ['headlines_df','time_df','link_df']
cnbc_df
```
