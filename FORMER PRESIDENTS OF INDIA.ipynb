# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

url = 'https://presidentofindia.nic.in/former-presidents.htm' 
page = requests.get(url)
page

president_soup = bs(page.content)
print(president_soup.prettify())
```

```python
name_president = []
for i in president_soup.find_all('img',class_="alignLeft"):
    name_president.append(i.get('title'))
#name_president
index = np.arange(1,len(name_president)+1)
name_president_df = pd.DataFrame(name_president, index=index)
name_president_df
```

```python
term = []
for j in president_soup.find_all('div', class_="presidentListing"):
    term.append(j.p.text.replace("Term of Office:", ''))
#term

index = np.arange(1,len(term)+1)
term_df = pd.DataFrame(term, index=index)
term_df
```

```python
presidents = pd.concat([name_president_df, term_df], axis=1)
presidents.columns = ['name', 'office_term']
presidents
```
