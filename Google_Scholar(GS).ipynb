```python
# Importing the libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page
```

```python
scholar = bs(page.content)
#print(scholar.prettify())
```

```python
publication = []
for i in scholar.find_all('td', class_="gsc_mvt_t"):
    publication.append(i.text)
publication
```

```python
h5_all = []
for h in scholar.find_all('td',class_="gsc_mvt_n"):
    h5_all.append(h.text)
h5_all
```

```python
h5_index =[]
h5_median = []
for x in range(len(h5_all)):
    if x%2 == 0:
        h5_index.append(h5_all[x])
    else:
        h5_median.append(h5_all[x])
```

# creating dataframe without using dictionaries

```python
index = np.arange(1,len(publication)+1)
print(index)
```

```python
publication_df = pd.DataFrame(publication, index=index)
print(publication_df)
```

```python
h5_index_df = pd.DataFrame(h5_index,  index=index)
print(h5_index_df)
```

```python
h5_median_df = pd.DataFrame(h5_median,  index=index)
print(h5_median_df)
```

# creating new series to set as rank

```python
scholars = pd.concat([publication_df, h5_index_df, h5_median_df], axis=1)
scholars.columns =['publication', 'h5_index', 'h5_median']
print(scholars)
```

```python
#creating rank column
scholars['rank'] = np.arange(1, len(scholars)+1)
scholars['rank']
```
