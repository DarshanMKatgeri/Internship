    a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.
    b) Top 10 women’s ODI Batting players along with the records of their team and rating.
    c) Top 10 women’s ODI all-rounder along with the records of their team and rating.

# a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
page = requests.get(url)
page
```

```python
soup = bs(page.content)
print(soup.prettify())
```

# details of rank-1 team (banner)
==================================================================

```python
team_banner = []
for team in soup.find_all('span', class_="u-hide-phablet"):
    team_banner.append(team.text)
team_banner
```

```python
team_banner_df = pd.DataFrame(team_banner)
index=np.arange(1,len(team_banner_df)+1)
team_banner_df = pd.DataFrame(team_banner, index=index)
team_banner_df.columns = ['teams']
team_banner_df
```

```python
matches_banner = []
for match in soup.find_all('td', class_="rankings-block__banner--matches"):
    matches_banner.append(match.text)
matches_banner
```

```python
point_banner = []
for point in soup.find_all('td', class_="rankings-block__banner--points"):
    point_banner.append(point.text)
point_banner
```

```python
rating_banner = []
for rating in soup.find_all('td', class_="rankings-block__banner--rating u-text-right"):
    rating_banner.append(rating.text.replace('\n','').replace(' ',''))
rating_banner
```

# creating dataframes of rank-1 (banner) team

```python
matches_banner_df = pd.DataFrame(matches_banner, index=[1]) 
matches_banner_df
```

```python
point_banner_df = pd.DataFrame(point_banner, index=[1]) 
point_banner_df
```

```python
rating_banner_df = pd.DataFrame(rating_banner, index=[1])
rating_banner_df
```

```python
banner = pd.concat([matches_banner_df,point_banner_df, rating_banner_df], axis=1)
banner.columns = ['matches','points','rating']
banner
```

# details of rank-2 onwards team
==================================================================

```python
matches_all = []
for match in soup.find_all('td', class_="table-body__cell u-center-text"):
    matches_all.append(match.text)
matches_all
```

```python
type(matches_all)
```

# separating matches and points into 2 lists

```python
matches= []
points = []
for i in range(len(matches_all)):
    if i%2 == 0:
        matches.append(matches_all[i])
    else:
        points.append(matches_all[i])
print("matches:",matches )
print("points:", points)
```

```python
matches_df = pd.DataFrame(matches)
index=np.arange(2,len(matches)+2)
matches_df = pd.DataFrame(matches, index=index)
matches_df.columns = ['matches']
matches_df
```

```python
points_df = pd.DataFrame(points)
index=np.arange(2,len(points)+2)
points_df = pd.DataFrame(points, index=index)
points_df.columns = ['points']
points_df
```

```python
ratings = []
for rate in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    ratings.append(rate.text)
ratings
index = np.arange(2,len(ratings)+2)
ratings_df = pd.DataFrame(ratings, index=index)
ratings_df.columns = ['rating']
ratings_df
```

```python
team_all = pd.concat([matches_df, points_df, ratings_df], axis=1)
team_all
```

```python
df_1 = pd.concat([banner, team_all],axis=0)
df_1
```

```python
tms_2 = []
for tm in soup.find_all('span', class_="u-show-phablet"):
    tms_2.append(tm.text)
tms_2
index = np.arange(1, len(tms_2)+1)
tms_2_df = pd.DataFrame(tms_2, index=index)
tms_2_df.columns = ['team_name']
tms_2_df
```

# Required result dataframe

```python
df_final = pd.concat([tms_2_df, df_1], axis=1)
df_final
```

# b) Top 10 women’s ODI Batting players along with the records of their team and rating.

```python
women_batsman_link = 'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting'
soup_women_odi_bat = requests.get(women_batsman_link)
soup_women_odi_bat
```

```python
women_odi_batsman = bs(soup_women_odi_bat.content)
print(women_odi_batsman.prettify())
```

```python
batsman_banner = []
for batsman in women_odi_batsman.find_all('div', class_="rankings-block__banner--name-large"):
    batsman_banner.append(batsman.text)
batsman_banner
```

```python
team_banner = []
for team in women_odi_batsman.find_all('div', class_="rankings-block__banner--nationality"):
    team_banner.append(team.text.replace('\n','').replace(' ',''))
team_banner
```

```python
rating_banner = []
for rating in women_odi_batsman.find_all('div', class_="rankings-block__banner--rating"):
    rating_banner.append(rating.text.replace('\n','').replace(' ',''))
rating_banner
```

# creating dataframe of rank-1 (banner)

```python
batsman_banner_df = pd.DataFrame(batsman_banner, index=[1])
batsman_banner_df.columns = ['name']
batsman_banner_df
```

```python
team_banner_df = pd.DataFrame(team_banner, index=[1])
team_banner_df.columns = ['country']
team_banner_df
```

```python
rating_banner_df = pd.DataFrame(rating_banner, index=[1])
rating_banner_df.columns = ['rating']
rating_banner_df
```

```python
banner_df = pd.concat([batsman_banner_df,team_banner_df,rating_banner_df], axis=1)
banner_df
```

# details of rank-2 onwards players

```python
batsman_all_names = []
for i in women_odi_batsman.find_all('td', class_="table-body__cell rankings-table__name name"):
    batsman_all_names.append(i.text.replace('\n',''))
print(batsman_all_names)
```

```python
batsman_all_country = []
for j in women_odi_batsman.find_all('span', class_="table-body__logo-text"):
    batsman_all_country.append(j.text)
print(batsman_all_country)
```

```python
batsman_all_rating = []
for k in women_odi_batsman.find_all('td',class_="table-body__cell rating"):
    batsman_all_rating.append(k.text)
print(batsman_all_rating)
```

# creating dataframes of rank-2 onwards batsman

```python
batsman_all_names_df = pd.DataFrame(batsman_all_names)
index = np.arange(2,len(batsman_all_names)+2)
batsman_all_names_df = pd.DataFrame(batsman_all_names, index=index)
batsman_all_names_df
```

```python
batsman_all_country_df = pd.DataFrame(batsman_all_country)
index = np.arange(2,len(batsman_all_country)+2)
batsman_all_country_df = pd.DataFrame(batsman_all_country, index=index)
batsman_all_country_df
```

```python
batsman_all_rating_df = pd.DataFrame(batsman_all_rating)
index = np.arange(2,len(batsman_all_rating)+2)
batsman_all_rating_df = pd.DataFrame(batsman_all_rating, index=index)
batsman_all_rating_df
```

# concating dataframe from rank-2 onwards 

```python
batsman_all = pd.concat([batsman_all_names_df, batsman_all_country_df, batsman_all_rating_df], axis=1)
batsman_all.columns = ['name','country','rating']
batsman_all
```

# concatenating banner and rank-2 onwards dataframes to get final required dataframe

```python
top_100 = pd.concat([banner_df, batsman_all], axis=0)
top_100
```

# top-10 batsman

```python
top_10_batsman = top_100.head(10)
top_10_batsman
```
