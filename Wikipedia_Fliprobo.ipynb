## Write a python program to display all the header tags from wikipedia.org.

```python
#importing required libraries
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
```

```python
#sending requests 
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
#creating object soup
soup = bs(page.content)
print(soup.prettify())
```

```python
header_tag = soup.find_all('span',class_='mw-headline')
header = []
for i in header_tag:
    header.append(i.text.upper())
header
```

```python
type(header)
```

```python
#creating dataframe
header_df = pd.DataFrame(header)

#setting index to start from 1
index = np.arange(1,len(header_df)+1)
header_df = pd.DataFrame(header, index=index)

#setting column title to header_tags
header_df.columns = ['header_tags']

#printing dataframe
header_df
```
